```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("retail (1).db")

with open("analysis_queries.sql") as f:
    script = f.read()

statements = [s.strip() for s in script.split(";") if s.strip()]
statements = [s for s in statements if any(
    line.strip() and not line.strip().startswith("--") for line in s.split("\n")
)]

for i, stmt in enumerate(statements, 1):
    print(f"--- Query {i} ---")
    display(pd.read_sql_query(stmt, conn))
```
